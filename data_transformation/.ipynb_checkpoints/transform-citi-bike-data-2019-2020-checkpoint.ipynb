## Data Cleansing 

- Merge Tables 
- View Data Types
- Check Null Values
- Change Column Info

```python
citi_bike_data = [citi_bike_jun19, citi_bike_jul19, citi_bike_aug19, 
                  citi_bike_sept19, citi_bike_oct19, citi_bike_nov19, 
                  citi_bike_dec19, citi_bike_jan20, citi_bike_feb20, 
                  citi_bike_mar20, citi_bike_apr20, citi_bike_may20, citi_bike_jun20]

citi_bike_merge = reduce(lambda left,right: pd.merge(left, right, how="outer"), citi_bike_data)
```

```python
citi_bike_merge.head()
```

```python
citi_bike_merge.info()
```

```python
citi_bike_merge.shape
```

```python
citi_bike_merge.isnull().sum()
```

```python
# If null values were found in merge, dropna would of been executed 
# citi_bike_merge.dropna(inplace=True)
```

```python
citi_bike_merge["gender"].replace({0: "unknown", 1: "male", 2: "female"}, inplace=True)
```

```python
citi_bike_merge.head()
```

## Save Clean Data Set

```python
citi_bike_clean = citi_bike_merge
```

```python
citi_bike_clean.to_csv('../data/citi_bike_data.csv')
```
